# Distribuições Amostrais

```python
import numpy as np
import matplotlib.pyplot as plt

# Função para criar distribuições amostrais das médias
def sample_means_distribution(data, column, sample_sizes, num_samples=1000):
    plt.figure(figsize=(16, 12))
    for i, size in enumerate(sample_sizes, 1):
        sample_means = [np.mean(np.random.choice(data[column], size)) for _ in range(num_samples)]
        plt.subplot(3, 1, i)
        plt.hist(sample_means, bins=20, color='skyblue', edgecolor='black', density=True)
        plt.title(f'Distribuição Amostral das Médias - Tamanho da Amostra: {size}')
        plt.xlabel('Média Amostral')
        plt.ylabel('Frequência')

    plt.tight_layout()
    plt.show()

# Executando a função para a coluna 'Market Value' com tamanhos de amostra variados
sample_sizes = [5, 30, 100]  # Definindo três tamanhos de amostra
sample_means_distribution(data, 'Market Value', sample_sizes)
```

# Intervalo de Confiança

```python
import numpy as np
import scipy.stats as stats

# Selecionando uma coluna para cálculo do Intervalo de Confiança
# Vamos usar a coluna 'Market Value' como amostra representativa

# Configurações iniciais
confidence_levels = [0.90, 0.95]  # Níveis de confiança de 90% e 95%
sample = data['Market Value'].dropna()  # Removendo valores ausentes, se houver
sample_mean = np.mean(sample)  # Média da amostra
sample_std = np.std(sample, ddof=1)  # Desvio padrão da amostra (com correção de Bessel)
n = len(sample)  # Tamanho da amostra

# Calculando e exibindo os Intervalos de Confiança
for confidence in confidence_levels:
    margin_of_error = stats.t.ppf((1 + confidence) / 2, n - 1) * (sample_std / np.sqrt(n))
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error
    print(f"Intervalo de Confiança de {int(confidence * 100)}%: ({lower_bound:.2f}, {upper_bound:.2f})")
```

```python
import numpy as np
import scipy.stats as stats

# Selecionando a coluna 'Goals' (número de gols) como amostra representativa
confidence_levels_goals = [0.90, 0.95]  # Níveis de confiança de 90% e 95%
sample_goals = data['Goals'].dropna()  # Removendo valores ausentes, se houver
sample_mean_goals = np.mean(sample_goals)  # Média da amostra
sample_std_goals = np.std(sample_goals, ddof=1)  # Desvio padrão da amostra (com correção de Bessel)
n_goals = len(sample_goals)  # Tamanho da amostra

# Calculando e exibindo os Intervalos de Confiança para 'Goals'
for confidence in confidence_levels_goals:
    margin_of_error_goals = stats.t.ppf((1 + confidence) / 2, n_goals - 1) * (sample_std_goals / np.sqrt(n_goals))
    lower_bound_goals = sample_mean_goals - margin_of_error_goals
    upper_bound_goals = sample_mean_goals + margin_of_error_goals
    print(f"Intervalo de Confiança de {int(confidence * 100)}% para 'Goals': ({lower_bound_goals:.2f}, {upper_bound_goals:.2f})")
```
